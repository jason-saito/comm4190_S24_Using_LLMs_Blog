<h1>How Does It Work?</h1>
In order to replicate music, the AI is trained on large data sets of pre-existing music. From this data set it analizes the various patterns, tempos, rythms, and lyrics associated with various genres of music.

<h1>Training the Model</h1>
Traditionally, when training a model, the model takes the data and makes a numerical vector. In order to allow the model to understand the various aspects that go into creating music, the model is trained with MIDI files that are structured in terms of notes, rhyhtm changes, and beats per minute. This MIDI file can be used to train the model and is considered a "natural language representation."

The Machine Learning algorithms then use the raw audio data for the entire song in the form of "sequential input vectors" that are commonly used in NLP. The model then is able to predict the next token for the song. 

<h1>Legal Implications of AI Music</h1>
With the rise of AI music, there is also a rise in copyright and licensing issues. As, when someone uses AI to imitate another artist's music, this can muddy the waters of who owns the music, the creator, the AI, or the artist themself.
